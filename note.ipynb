```python
import plotly.express as px
import pandas as pd

# Lista de tickers a monitorar
monitor_tickers = ['GOOGL', 'MSFT', 'AAPL']

# Leitura do DataFrame
df_filtered = pd.read_csv('data/stocks.csv')

# Garantir que a coluna 'Datetime' esteja no formato datetime
df_filtered['Datetime'] = pd.to_datetime(df_filtered['Datetime'])

# Filtrar apenas os tickers monitorados
df_filtered = df_filtered[df_filtered["Ticker"].isin(monitor_tickers)]

# Calcular a variação percentual para cada ticker
def calcular_variacao_percentual(df):
    df = df.sort_values(by="Datetime")  # Ordenar por data
    df["Percentual_Variacao"] = (df["Close"] / df["Close"].iloc[0] - 1) * 100  # Variação percentual
    return df

# Aplicar para cada ticker de maneira mais explícita, sem usar groupby com apply
df_percentual = pd.concat([calcular_variacao_percentual(df_filtered[df_filtered['Ticker'] == ticker]) 
                           for ticker in monitor_tickers])

# Criar o gráfico
fig = px.line(
    df_percentual,
    x="Datetime",
    y="Percentual_Variacao",
    color="Ticker",  # Diferencia as linhas por ticker
    title="Variação Percentual do Preço de Fechamento",
    labels={"Percentual_Variacao": "Variação Percentual (%)", "Datetime": "Data"}
)

# Mostrar o gráfico
fig.show()
```
